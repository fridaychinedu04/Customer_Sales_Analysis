### **INTRODUCTION**
This report summarizes key business insights obtained from an in-depth analysis of sales, customer behaviour, product performance, and operational efficiency. Using Excel Pivot Tables, various dimensions of the dataset, such as product categories, ratings, shipping charges, customer demographics, and temporal trends, were examined to uncover patterns that can support data-driven decision-making.
The purpose of this report is to highlight meaningful trends and relationships while presenting results in a structured and professional format suitable for stakeholders, management, or academic review.

---
#### **OBJECTIVES**
The primary objectives of this analysis are as follows:
- Evaluate differences in average ratings across product categories
- Determine whether a relationship exists between shipping charges and total sales
- Identify the age group with the highest average sales per order
- Compare ratings for international versus local orders
- Determine which locations consistently perform well in both quantity and revenue
- Assess which product category displays the highest month-over-month sales growth
- Compare buying behaviour between male and female buyers in terms of order value and quantity
- Analyze average sales per unit within product categories and its relationship to quantity sold
- Identify the month with the greatest sales increase relative to the previous month
- Examine potential patterns between order ratings and the presence of shipping charges

---
##### **KEY FINDINGS**
1. ***How does the average rating differ across product categories?***
        - **Other:** **3.582**
        - **Clothing:**  **3.499**
        - **Ornaments:**  **3.446**
"Other" products receive the highest average rating; "Ornaments" the lowest. 
2. ***Is there correlation between shipping charges and total sales?***
A moderate positive correlation - orders with higher shipping charges tend to be associated with higher total sales (possibly because larger/expensive orders incur shipping fees). Correlation is not extremely strong, so, shipping charges are only one of several factors tied to order size.
3. ***Which age group provides the highest sales per order on average***
Buyers aged **26-30** generate the highest average order value - good target segment for promotions/upsells.
4. ***Do international orders tend to have higher or lower ratings than local ones?***
        - **International(Yes):** Average Rating = **3.509**
        - **Local(No):** Average Rating = **3.495**
International orders have a slightly average rating (0.014 difference) - essentially similar, no large difference.
5. ***Are certain locations consistently high performers across both quantity and revenue?***
A small set of locations consistently contribute both high volume and high revenue - these are prime candidates for targeted logistics and marketing investments
6. ***Which product category shows the highest sales growth month-over-month?***
Clothing showed the largest single month-over-month percent growth observed in the dataset: **May 2024** with **~23.59% month-over-month increase** for Clothing. In absolute total sales across months, **May 2024** was a strong month(also the month with the largest overall monthly increase).
Clothing experienced the biggest proportional jump in a single month; investigate promotions, campaigns or seasonality in May to replicate success.
7. ***Do male and female buyers differ in average order value or quantity?***
       - **Female:** Average Order Value = **116.11**, Average Quantity per Order = **1.65**
       - **Male:** Average Order Value = **115.71**, Average Quantity per Order = **1.68**
Differences are negligible. average order value and quantity per order are very similar between male and female buyers.
8. ***What is the average sales per unit in each product category and how does that relate to total quantity sold?***
        - **Clothing:** Average Sales per Unit = **104.57**, Total Quantity Sold = **6097**
        - **Ornaments:** Average Sales per Unit = **41.20**, Total Quantity Sold = **3768**
        - **Other:** Average Sales per Unit = **24.98**, Total Quantity Sold = **2469**
**Clothing** sells at a much higher price per unit and also has the largest quantity sold - strong revenue driver. **Ornaments** and **Other** have lower unit prices; **Ornaments** sells more units than **Other** but still far less average unit price than **Clothing**.
9. ***Which month had the highest increase in sales compared to the previous month?***
May 2024 is the best month for growth - dig into promotions, inventory, or external factors that led to this jump.
10. ***Is there pattern between order rating and the presence of shipping charges?***
There is no strong negative relationship between having shipping charges and rating - customers paying shipping charges do not rate orders noticeably lower

---
##### **RECOMMENDATION**
- Target age **26-30** with targeted offers highest average order value
- **Clothing** is the most valuable category (high average price and strong month-over-month growth in May) - prioritize inventory, marketing, and merchandising here. 
- **Shipping fees** shows a moderate positive correlation with total sales - consider bundling or shipping strategies for large orders (free-shipping thresholds).
- **International customers** have comparable (slightly higher) ratings - mainatian international service quality and shipping reliability.
- **Top locations** that rank high in both quantity and revenue should be prioritized for faster logistics, localized marketing campaigns, and stocking.
- **Gender differences** are negligible; use other segmentation (age, location, category) for targeting.
- **Monthly spikes (May)** merit further investigation - replicate successful tactics (promotions, partnerships, seasonality).

---
###### **CONCLUSION**
This analysis highlights meaningful insights across product performance, customer behaviour, and sales trends. These findings can support strategic decisions such as optimizing inventory, refining marketing strategies, enhancing customer experience, and identifying growth opportunities.
Excel Pivot Table proved efficient for summarizing large datasets, calculating averages, identifying correlations, and performing month-over-month comparisons. By applying these analytical techniques periodically, organizations can continuously monitor performance and make informed decisions that drive business growth.
